Objective - For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it
graphically.

```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.DataFrame(df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm","PetalWidthCm"]]) #independent
y = df["Species"] #target
```

```python
#dividing the dataset into train & test
from sklearn.model_selection import train_test_split
xtrain , xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
#building the decision tree clasifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
y_pred = dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,y_pred))
```

```python
features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm","PetalWidthCm"]
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
from sklearn.tree import plot_tree #visualizing the decision tree
from sklearn import tree
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10), dpi=100)
tree.plot_tree(dt, feature_names=features, class_names=targets, filled=True)
fig.savefig('Decision_tree.png')
```
